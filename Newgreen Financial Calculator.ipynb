```python
def Investment_calc():
    init_amt = int(input("Starting Amount: "))
    time = int(input("Time in years: "))
    interest_rate = int(input("Interest Rate Percentage: "))
    Compound = input("Compounded Annually or Monthly? Enter Annual or Month: ")
    add_cont = int(input("Add any additional contribution amount: "))
    interest_amt = 0
    temp = init_amt
    if Compound == "Annual" or Compound == "annual":
        for i in range(time):
            temp += add_cont
            interest_amt = temp * (interest_rate / 100)
            temp += interest_amt
            print(f"Investment value at end of year {i + 1} is {round(temp, 2)}")
    elif Compound == "Month" or Compound == "month":
        time = time * 12
        for i in range(time):
            temp += add_cont
            interest_amt = temp * ((interest_rate / 100) / 12) 
            temp += interest_amt
            print(f"Investment value at end of month {i + 1} is {round(temp, 2)}")
```

```python
Investment_calc()
```
